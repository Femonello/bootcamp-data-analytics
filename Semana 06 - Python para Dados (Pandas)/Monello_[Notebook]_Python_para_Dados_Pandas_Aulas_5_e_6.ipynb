```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

df_concurso = pd.read_csv('/content/drive/MyDrive/Womakers - Datasets/dados_concurso.csv', sep=',')
df_concurso.head()
```

```python
#Quantas pessoas são do estado do Piauí?

pessoas_do_piaui= df_concurso[df_concurso['Estado']=='PI']['Estado'].count()
print(f'Pessoas do Piaui: {pessoas_do_piaui}')
```

```python
# Quantas do ES são so sexo feminino?

pessoas_do_es_feminino= df_concurso[(df_concurso['Estado']=='ES') & (df_concurso['Sexo']=='F')]['Estado'].count()
print(f'Pessoas do Piaui: {pessoas_do_es_feminino}')
```

```python
# Quantas pessoas nasceram antes de 1995?

# Converter a coluna "Data de Nascimento" para o tipo datetime (se ainda não estiver)
df_concurso['Data de Nascimento'] = pd.to_datetime(df_concurso['Data de Nascimento'], errors='coerce')

# Filtrar as pessoas nascidas antes de 1995
nasc_antes_95 = df_concurso[df_concurso['Data de Nascimento'].dt.year < 1995]['Data de Nascimento'].count()

nasc_antes_95
```

```python
# Retorne os dados dos incritos com as colunas: nº inscrição, nome e idade.
from datetime import datetime

# Converter a coluna "Data de Nascimento" para o tipo datetime (se ainda não estiver)
df_concurso['Data de Nascimento'] = pd.to_datetime(df_concurso['Data de Nascimento'], errors='coerce')

# Calcular a idade
hoje = pd.Timestamp('now')
df_concurso['Idade'] = (hoje.year - df_concurso['Data de Nascimento'].dt.year) - (
    (hoje.month < df_concurso['Data de Nascimento'].dt.month) |
    ((hoje.month == df_concurso['Data de Nascimento'].dt.month) & (hoje.day < df_concurso['Data de Nascimento'].dt.day))
).astype(int)

df_concurso[['Número de Inscrição', 'Nome', 'Idade']]
```
