```python
df_concurso = pd.read_csv('1EdHY4mNQ9WMpUkbCtMK_dados_concurso.csv')
df_concurso.head()
```

```python
# Qual a média de idade por região?

from datetime import datetime

# Converter a coluna de data de nascimento para datetime
df_concurso['Data de Nascimento'] = pd.to_datetime(df_concurso['Data de Nascimento'])

# Calcular a idade
hoje = pd.Timestamp('now')
df_concurso['Idade'] = (hoje.year - df_concurso['Data de Nascimento'].dt.year) - (
    (hoje.month < df_concurso['Data de Nascimento'].dt.month) |
    ((hoje.month == df_concurso['Data de Nascimento'].dt.month) & (hoje.day < df_concurso['Data de Nascimento'].dt.day))
).astype(int)

# Agrupar por região e calcular a média de idade
media_idade_por_regiao = df_concurso.groupby('Estado')['Idade'].mean()

print(media_idade_por_regiao)


```

```python
#Qual é a porcentagem de pessoas com deficiência?

porcentagem_com_deficiencia = (df_concurso['Deficiência'] == 'Sim').mean() * 100

print(f"A porcentagem de pessoas com deficiência é: {porcentagem_com_deficiencia:.2f}%")
```

```python
#Quantas pessoas são do estado do Piauí?

pessoas_do_piaui = df_concurso[df_concurso['Estado'] == 'PI']

quantidade_pessoas_piaui = len(pessoas_do_piaui)

quantidade_pessoas_piaui
```
