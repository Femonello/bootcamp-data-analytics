## **#2**

Qual é a média, a moda e a mediana de horas de sono para cada uma

```python
df_media_moda_mediana = df.groupby('Profissão')['Duração do sono'].agg([('Media_Sono', np.mean), ('Mediana_Sono', np.median), ('Moda_sono', pd.Series.mode)])

df_media_moda_mediana
```

## **#3**

Das pessoas que atuam com engenharia de software qual a
porcentagem de obesos?

```python

# Calculando o total de pessoas que atuam como Engenheiras de Software
total_eng_software = len(df[df['Profissão'] == 'Eng. de Software'])

# Calculando a quantidade de pessoas que atuam como Engenheiras de Software e são obesas
eng_software_obesos = len(df[(df['Profissão'] == 'Eng. de Software') & (df['Categoria IMC'] == 'Obesidade')])

# Calculando a porcentagem
pp_eng_software_obesos = (eng_software_obesos / total_eng_software) * 100

print(f'{pp_eng_software_obesos:.2f}% dos Engenheiros de Software estão categorizados com obesidade.')
```

## **#4**

De acordo com os dados, advogar ou ser representante de vendas faz
você dormir menos? (Use o método ‘isin’, considere a média)

```python
df['Profissão'].unique() #Extraindo lista de profissões para usar texto
```

```python
#criando lista das profissões selecionadas
adv_e_vendas =['Advogado(a)', 'Representante de Vendas']

# Criando um novo DataFrame filtrando as profissões da lista usando o método 'isin
df_adv_vendas = df[df['Profissão'].isin(adv_e_vendas)]
df_adv_vendas.head()
```

```python
# Média de sono por profissão

media_de_sono_por_profissao = df_adv_vendas.groupby('Profissão')['Duração do sono'].mean()
media_de_sono_por_profissao
```

```python
# Média de sono das duas profissões

media_de_sono = df_adv_vendas['Duração do sono'].mean()
media_de_sono
```

```python
# Média de sono de todos os registros

media_de_sono_geral = df['Duração do sono'].mean()
media_de_sono_geral
```

```python
# comparando as médias por profissão

variacao_percentual = ((media_de_sono_por_profissao - media_de_sono_geral) / media_de_sono_geral) * 100
variacao_percentual
```

```python
# Comparando as  médias gerais

variacao_percentual = ((media_de_sono - media_de_sono_geral) / media_de_sono_geral) * 100
variacao_percentual
```

### **Conclusão do resultado**

A análise dos dados revela que a média de duração do sono das duas profissões avaliadas apresenta uma variação de aproximadamente 3,08 p.p. acima da média geral.

Ao observar as profissões separadamente, percebe-se que os Representantes de Vendas dormem, em média, 17% menos do que a média geral.

No entanto, esses dados não permitem afirmar que a profissão de Representante de Vendas seja a principal causa da menor duração do sono. Outros fatores, como estilo de vida, carga horária de trabalho e hábitos individuais, podem influenciar esse resultado. Para uma conclusão mais precisa, seria necessária uma análise mais aprofundada, considerando variáveis adicionais.

## **8**

Quem tem maior pressão sanguínea média, homens ou mulheres?
(Considere a média)

```python
df['Pressão Sangüínea'].info() #verificando o tipo da coluna de pressão
```

```python
# Pressão é medida em Sistolica e Diastolica

# Separar a medição em duas colunas adicionando-as no DataFrame
df[['Sistolica', 'Diastolica']] = df['Pressão Sangüínea'].str.split('/', expand=True)

# Convertendo as colunas para números inteiros
df[['Sistolica', 'Diastolica']] = df[['Sistolica', 'Diastolica']].astype(int)

df.head()
```

```python
# Calculando a média ara cada Gênero
media_pressao = df.groupby('Gênero')[['Sistolica', 'Diastolica']].mean()
media_pressao
```

### **Conclusão**

Em média, as mulheres do conjunto de dados analisado têm pressão arterial mais alta do que os homens.

## **#10**

Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
df['Frequência cardíaca'].info()
```

```python
grupo_FC_acima_70 = df[df['Frequência cardíaca'] > 70]['Passos diários'].mean()
grupo_FC_acima_70
```

```python
grupo_FC_abaixo_70 = df[df['Frequência cardíaca'] <= 70]['Passos diários'].mean()
grupo_FC_abaixo_70
```

```python
variacao_percentual = ((grupo_FC_acima_70 - grupo_FC_abaixo_70) / grupo_FC_abaixo_70) * 100
variacao_percentual
```

```python
if variacao_percentual <= 0:
  print(f'Pessoas com frequências cardíacas acima de 70 dá, em média, {variacao_percentual*-1:.2f}%"  menos passos do que o grupo com frequência cardíaca menor ou igual a 70.')
else:
  print(f'Pessoas com frequências cardíacas acima de 70 dá, em média, {variacao_percentual:.2f}%"  mais passos do que o grupo com frequência cardíaca menor ou igual a 70.')
```
