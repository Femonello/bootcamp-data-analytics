```python

df_veiculos = pd.read_csv('UAThHE0RcyyZb6QkRNqp_carros.csv')
df_veiculos.head(10)
```

```python
 #print as colunas e index do dataframe
 print(df_veiculos.columns)
 print(df_veiculos.index) #O RangeIndex é um tipo de índice padrão usado pelo Pandas quando você cria um DataFrame sem especificar um índice personalizado.
                          #É como uma contagem automática de linhas, funcionando de forma parecida com o range() do Python.
                          #Saída:
                             #start = onde começa, primeira linha do Dataframe;
                             #stop = quantiadde de linhas do DataFrame. O resultado é excludente, ou seja, se o retorno for 27, o ínidce máximo é 26
                             #stop = unidades em que o índice avança (de 1 em 1, de 2 em 2 etc)
```

```python
#Faça um sort decrescente no dataframe pelo estado com maior quantidade de carros

df_veiculos.sort_values(by='Quant Carros', ascending=False)
```

```python
#Descubra qual a proporção de carro para a população de cada estado e insira em uma nova coluna

var_proporcao = df_veiculos['Quant Carros'] / df_veiculos['População do Estado']
var_proporcao
```

```python
#inserindo a proporção em uma coluna no Dataframe
df_veiculos['Proporção'] = var_proporcao
df_veiculos
```

```python
#Faça um dataframe com Estado e proporção de carro por habitante

df_novo = df_veiculos[['Estado', 'Proporção']]
df_novo
```
